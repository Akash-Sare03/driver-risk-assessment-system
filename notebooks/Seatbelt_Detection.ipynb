```python
# Rename Class Folders

import os

base_path = "Seat_Belt2"

rename_map = {
    "Seat_Belt": "seatbelt",
    "WithoutSeat_Belt": "no_seatbelt"
}

for split in ["Train", "Test"]:
    split_path = os.path.join(base_path, split)
    for old, new in rename_map.items():
        old_path = os.path.join(split_path, old)
        new_path = os.path.join(split_path, new)
        if os.path.exists(old_path):
            os.rename(old_path, new_path)
```

```python
# Create Validation Set from Train

import random
import shutil

random.seed(42)

base_path = "Seat_Belt2"
train_path = os.path.join(base_path, "Train")
val_path = os.path.join(base_path, "val")

os.makedirs(val_path, exist_ok=True)

for class_name in ["seatbelt", "no_seatbelt"]:
    train_class_dir = os.path.join(train_path, class_name)
    val_class_dir = os.path.join(val_path, class_name)

    os.makedirs(val_class_dir, exist_ok=True)

    images = os.listdir(train_class_dir)
    random.shuffle(images)

    split_size = int(0.2 * len(images))
    val_images = images[:split_size]

    for img in val_images:
        src = os.path.join(train_class_dir, img)
        dst = os.path.join(val_class_dir, img)
        shutil.move(src, dst)
```

```python
from torchvision import transforms
from torch.utils.data import random_split, Dataset, DataLoader
from torchvision.datasets import ImageFolder
```

```python
# Transforms (ImageNet-style)
# MobileNet expects:

'''Image size: 224  224

ImageNet normalization'''

train_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

val_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
```

```python
# Dataset & DataLoader


train_ds = ImageFolder("Seat_Belt2/Train", transform=train_transforms)
val_ds   = ImageFolder("Seat_Belt2/val", transform=val_transforms)

train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)
val_loader   = DataLoader(val_ds, batch_size=32, shuffle=False)

print(train_ds.class_to_idx)
```

```python
# Load MobileNet (Transfer Learning)

import torch
import torch.nn as nn
from torchvision import models

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model = models.mobilenet_v2(pretrained=True)

#Freeze Backbone
for param in model.parameters():
    param.requires_grad = False
```

```python
# Replace Final Layer

model.classifier[1] = nn.Linear(model.last_channel, 2)
model = model.to(device)
```

```python
# Loss Function & Optimizer

import torch.optim as optim

criterion = nn.CrossEntropyLoss()

optimizer = optim.Adam(
    model.classifier.parameters(),
    lr=0.001
)
```

```python
# Training Loop 

from tqdm import tqdm

epochs = 5

for epoch in range(epochs):
    model.train()
    train_correct = 0
    train_total = 0
    train_loss = 0

    for images, labels in tqdm(train_loader):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        train_loss += loss.item()
        _, preds = torch.max(outputs, 1)
        train_correct += (preds == labels).sum().item()
        train_total += labels.size(0)

    train_acc = train_correct / train_total

    # Validation
    model.eval()
    val_correct = 0
    val_total = 0

    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, preds = torch.max(outputs, 1)
            val_correct += (preds == labels).sum().item()
            val_total += labels.size(0)

    val_acc = val_correct / val_total

    print(f"Epoch [{epoch+1}/{epochs}] "
          f"Train Acc: {train_acc:.2f} | Val Acc: {val_acc:.2f}")
```

```python
# Evaluation on TEST Dataset

test_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

test_ds = ImageFolder("Seat_Belt2/Test", transform=test_transforms)
test_loader = DataLoader(test_ds, batch_size=32, shuffle=False)
```

```python
# Evaluate Test Accuracy

model.eval()
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, preds = torch.max(outputs, 1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)

test_acc = correct / total
print(f"Test Accuracy: {test_acc:.2f}")
```

```python
# Inference Code

from PIL import Image
import torch.nn.functional as F

class_names = test_ds.classes  # ['no_seatbelt', 'seatbelt']

def predict_seatbelt(image_path, model):
    model.eval()

    image = Image.open(image_path).convert("RGB")
    image = test_transforms(image)
    image = image.unsqueeze(0).to(device)

    with torch.no_grad():
        outputs = model(image)
        probs = F.softmax(outputs, dim=1)
        confidence, pred = torch.max(probs, 1)

    return {
        "label": class_names[pred.item()],
        "confidence": round(confidence.item(), 3)
    }
```
